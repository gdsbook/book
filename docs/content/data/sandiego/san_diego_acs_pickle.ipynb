## Replace NAN values with column mean

```python
for col in df_copy.columns:
    print(col)
    try:
        df_copy[col].fillna(((df_copy[col]).mean()), inplace=True)
    except:
        print(col)
```

```python
import pysal as ps
```

```python
df_copy.columns
```

```python
w_rook = ps.weights.Rook.from_dataframe(df_copy)
```

```python
w_rook.n
```

```python
w_rook.islands
```

```python
w_k4 = ps.weights.KNN.from_dataframe(df_copy)
```

```python
w_k4.islands
```

```python
w_k4.transform = 'r'
```

```python
year_built = df_copy['B25077_001E']
```

```python
yb_mi = ps.Moran(year_built, w_k4)
```

```python
yb_mi.p_norm
```

```python
yb_mi.I
```

```python
yb_mi.EI
```

```python
import numpy as np
np.random.seed(12345)
yb_lmi = ps.Moran_Local(year_built, w_k4)
```

```python
(yb_lmi.p_sim < 0.05).sum()
```
